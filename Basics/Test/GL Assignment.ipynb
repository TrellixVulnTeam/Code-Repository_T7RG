```python
# Reading the dataframe 
import pandas as pd

df=pd.read_csv('Wholesale+Customers+Data.csv')
df.head()
```

## Question 1 
Use methods of descriptive statistics to summarize data. Which Region and which Channel spent the most? Which Region and which Channel spent the least?


```python
# Decriptive statistics
df.describe()
```

```python
df.columns
```

```python
# Creating a new column to find the total spend
df['Total spend']=df['Fresh']+df['Milk']+df['Grocery']+df['Frozen']+df['Detergents_Paper']+df['Delicatessen']
```

```python
# Using groupby to get the total spend of each channels
channel=df[['Total spend','Channel']].groupby(by='Channel').sum()
channel
```

```python
# Using groupby to get the total spend of each region
region=df[['Total spend','Region']].groupby(by='Region').sum()
region
```

```python
# Plotting a graph for the spend of different channels
import matplotlib.pyplot as plt
plt.title('Total Spend of different channels')
plt.bar(channel.index,channel['Total spend'])
```

```python
# Plotting a graph for the spend of different regions
import matplotlib.pyplot as plt
plt.title('Total Spend of different regions')
plt.bar(region.index,region['Total spend'])
```

### In channels, hotels has the maximum spend and in regions Others has the maximum spent, followed by Lisbon and then by Oporto
