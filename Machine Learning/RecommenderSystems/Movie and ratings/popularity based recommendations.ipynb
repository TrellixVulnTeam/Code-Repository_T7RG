# Popularity based recommender system
This takes into considerdation the ratings and number of people who has rated a movie. <br>
T

```python
import numpy as np  
import pandas as pd
```

```python
ratings_data = pd.read_csv("ratings.csv")  
ratings_data.head() 
```

```python
movie_names = pd.read_csv("movies.csv")  
movie_names.head()  
```

```python
movie_data = pd.merge(ratings_data, movie_names, on='movieId')  
```

```python
movie_data.head()  
```

```python
movie_data.groupby('title')['rating'].mean().head()  
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()  
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()  
```

```python
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean()) 
```

```python
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())  
```

```python
ratings_mean_count.head()  
```

```python
ratings_mean_count['Normalised Rating']=ratings_mean_count['rating_counts']/ratings_mean_count['rating']
```

```python
ratings_mean_count.head()
```

# Ensuring the the number of people does not affect the rating by Normalising the same

```python
ratings_mean_count.sort_values(by='Normalised Rating',ascending=False)
```
