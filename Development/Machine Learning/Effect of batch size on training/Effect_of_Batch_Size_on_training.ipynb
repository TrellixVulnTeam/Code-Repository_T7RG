```
import tensorflow as tf
#oading the data from inbuilt libraries
(trainX, trainY),(testX, testY) = tf.keras.datasets.mnist.load_data()
#OneHotEncoding the data using keras.util.tocategorical function
trainY=tf.keras.utils.to_categorical(trainY,num_classes=10)
testY=tf.keras.utils.to_categorical(testY,num_classes=10)
# Bulding the graph using sequential function
```

```
# With SGD Optimizer
```

```
def create_model():


  model=tf.keras.Sequential()

  #Reshaping the 28*28 image to a 1D vector for processing in the layers

  model.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))
  #Adding layers to the pipeline

  model.add(tf.keras.layers.BatchNormalization())

  model.add(tf.keras.layers.Dense(200, activation='sigmoid'))
  model.add(tf.keras.layers.Dropout(0.3))

  model.add(tf.keras.layers.Dense(100,activation='relu'))
  model.add(tf.keras.layers.BatchNormalization())

  model.add(tf.keras.layers.Dense(60,activation='sigmoid'))
  model.add(tf.keras.layers.Dense(30,activation='sigmoid'))

  model.add(tf.keras.layers.Dense(10,activation='softmax'))

  #Optimizer and Loss

  model.compile(optimizer='sgd',loss='mse',metrics=['accuracy'])

  #Summary of the created model
  # model.summary()
  return model

```

```
def create_model_1():


  model=tf.keras.Sequential()

  #Reshaping the 28*28 image to a 1D vector for processing in the layers

  model.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))
  #Adding layers to the pipeline

  model.add(tf.keras.layers.BatchNormalization())

  model.add(tf.keras.layers.Dense(200, activation='sigmoid'))
  model.add(tf.keras.layers.Dropout(0.3))

  model.add(tf.keras.layers.Dense(100,activation='relu'))
  model.add(tf.keras.layers.BatchNormalization())

  model.add(tf.keras.layers.Dense(60,activation='sigmoid'))
  model.add(tf.keras.layers.Dense(30,activation='sigmoid'))

  model.add(tf.keras.layers.Dense(10,activation='softmax'))

  #Optimizer and Loss

  model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])

  #Summary of the created model
  # model.summary()
  return model

```

```
time1=[]
train_accuracy=[]
test_accuracy=[]
histories=[]
```

```
#Training with batch size =32
import time
import matplotlib.pyplot as plt
start=time.time()
model=create_model()
history=model.fit(trainX,trainY,batch_size=32,validation_data=(testX,testY),epochs=20)
print('Training completed in ',time.time()-start,' seconds')
train_accuracy.append(history.history['accuracy'][-1])
test_accuracy.append(history.history['val_accuracy'][-1])
time1.append(time.time()-start)
histories.append(history)
```
