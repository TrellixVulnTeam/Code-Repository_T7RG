```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 22 17:10:31 2021
Modified on Nov 2
@author: Sandheep Gopinath
"""

class myNode():
    
    def __init__(self,val):
        self.next=None
        self.value=val
        self.previous=None
        
class myList():

    def __init__(self):
        self.first=None
        self.length=0
        self.previous=None
    
    def add(self,value):
        node=myNode(value)
        if self.length==0:
            self.first=node
            self.length+=1
            self.previous=None
        else:
            node.next=self.first
            self.first=node
            self.length+=1
            self.first.previous=node


    def scan(self):
        listOfValues=[]
        if self.first==None:
            print('No elements')
        else:
            start=self.first
            for i in range(self.length):
                listOfValues.append(start.value)
                start=start.next
        return listOfValues
    
    def delete(self,val):
        if self.first==None:
            print('No elements')
        else:
            start=self.first
            previous=self.first
            for i in range(self.length):
                if start.value==val:
                    previous.next=start.next
                    self.length-=1
                previous=start
                start=start.next
    
    
    
l=myList()
l.add(5)
l.add(10)
l.add(15)

print(l.scan())

l.delete(5)
print(l.scan())
```
